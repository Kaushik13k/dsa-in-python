### Largest Rectangular Area in a Histogram using Stack

```python
histogram = [6, 2, 5, 4, 5, 1, 6]
stack = []
output_left = [-1] * len(histogram)
output_right = [len(histogram)] * len(histogram)

# Calculate the left boundaries for each bar
for i in range(len(histogram)):
    while stack and histogram[i] <= histogram[stack[-1]]:
        stack.pop()

    if stack:
        output_left[i] = stack[-1]

    stack.append(i)

# Clear the stack
stack = []

# Calculate the right boundaries for each bar
for i in range(len(histogram) - 1, -1, -1):
    while stack and histogram[i] <= histogram[stack[-1]]:
        stack.pop()

    if stack:
        output_right[i] = stack[-1]

    stack.append(i)

# Calculate the width of each rectangle
width = [0] * len(output_right)
for i in range(len(output_right)):
    width[i] = output_right[i] - output_left[i] - 1

# Calculate the maximum area
max_area = 0
for i in range(len(histogram)):
    max_area = max(max_area, width[i] * histogram[i])

print(max_area)
```

```python
def largestRectangleArea(heights):
    stack = []
    max_area = 0

    for i in range(len(heights)):
        while stack and heights[i] <= heights[stack[-1]]:
            height = heights[stack.pop()]
            width = i if not stack else i - stack[-1] - 1
            max_area = max(max_area, height * width)

        stack.append(i)

    while stack:
        height = heights[stack.pop()]
        width = len(heights) if not stack else len(heights) - stack[-1] - 1
        max_area = max(max_area, height * width)

    return max_area

# Example usage:
histogram = [6, 2, 5, 4, 5, 1, 6]
max_area = largestRectangleArea(histogram)
print(max_area)
```
